## Import Modules

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Function to detect faces

```python
def detect_faces(test_img, cascade = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml')):
    cpy_img = test_img.copy()
    g_img = cv2.cvtColor(cpy_img, cv2.COLOR_BGR2GRAY)
    rect_face = cascade.detectMultiScale(g_img, scaleFactor=1.1, minNeighbors=5)
    for (x, y, w, h) in rect_face:
        cv2.rectangle(cpy_img, (x, y), (x+w, y+h), (58, 255, 111), 5)
    return cpy_img
```

## Function to count the number of faces

```python
def face_count(test_img, cascade = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml')):
    cpy_img = test_img.copy()
    g_img = cv2.cvtColor(cpy_img, cv2.COLOR_BGR2GRAY)
    rect_faces = cascade.detectMultiScale(g_img, scaleFactor = 1.2, minNeighbors = 5);
    return len(rect_faces)
```

## Testing the function created

```python
test_image1 = cv2.imread('123.jpg')

faces = detect_faces(test_image1,cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml'))

no_of_faces = face_count(test_image1,cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml'))

print('Faces found: ', no_of_faces)

plt.imshow(cv2.cvtColor(faces, cv2.COLOR_BGR2RGB))
```

```python
cv2.imwrite("image1.jpg",faces)
```

```python
test_image2 = cv2.imread('456.jpg')

faces = detect_faces(test_image2,cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml'))

no_of_faces = face_count(test_image2,cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml'))

print('Faces found: ', no_of_faces)

plt.imshow(cv2.cvtColor(faces, cv2.COLOR_BGR2RGB))
```

## Saving the output file

```python
cv2.imwrite("image2.jpg",faces)
```
